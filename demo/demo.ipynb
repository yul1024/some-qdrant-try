```python
from langchain_qdrant import QdrantVectorStore
```

```python
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams
```

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embedding_model = OllamaEmbeddings(model='nomic-embed-text:v1.5')
```

```python
embedding_model.embed_query("haha")
```

```python
client = QdrantClient(":memory:")
```

```python
client.create_collection(
    collection_name="demo",
    vectors_config=VectorParams(
        size=768,
        distance=Distance.COSINE,
    )
)
```

```python
vector_store = QdrantVectorStore(
    client=client,
    collection_name="demo",
    embedding=embedding_model,
)
```

```python
vector_store.add_texts(texts=['doc1', 'doc2'])
```

```python
vector_store.similarity_search('haha')
```

```python
vector_store.max_marginal_relevance_search('haha')
```
